##  Fundamentos de Probabilidad 

### Definici贸n de Probabilidad

La probabilidad es una funci贸n matem谩tica que asigna, a ciertos subconjuntos de un espacio muestral, un n煤mero entre 0 y 1, que representa una medida del grado de certeza (o incertidumbre) con que ocurre un evento.

Formalmente, se define un *espacio de probabilidad* como una triada definida como $(\Omega,\mathcal{F}, P)$ donde $\Omega$ es el espacio muestral , es decir, el conjunto de todos los posibles resultados de un experimento aleatorio. $\mathcal{F}$ es una $\sigma$-谩lgebra sobre $\Omega$, es decir, una colecci贸n de subconjjntos de $\Omega$ (eventos) que cumple las siguientes propiedades : 

* $\Omega\in\mathcal{F}$
  
* Si $A\in\mathcal{F}$, entonces $A^c \in\mathcal{F}$

* Si $A_1, A_2,\cdots, A_n \in\mathcal{F}$, entones $\bigcup^{\infty}_{i=1} A_i \in\mathcal{F}$
  
Adem谩s, $P:\mathcal{F}\to [0,1]$ es una medida de probabilidad, una funci贸n que satisface los *axiomas de Kolmogorov*. 


### Axiomas de Kolm贸gorov

Sean $A, B, A_i \in\mathcal{F}$, entonces

* **Axioma 1:** No negatividad , $P(A)\geq 0$
* **Axioma 2:** Normalizacion, $P(\Omega)=1$
* **Axioma 1:** Aditividad Numerable, si los eventos son disjuntos (es decir $A_i \cap A_j = 0$ si $i\neq j$), entonces : 

\begin{equation*}
P\left( \bigcup^{\infty}_{i=1} A_i\right) = \sum_{i=1}^{\infty}P(A_i)
\end{equation*}

A partir de los axiomas anteriores, se derivan reglas 煤tiles como las siguientes : 

* **Complemento** $P(A^c) = 1 - P(A)$
  
* **Uni贸n** $P(A\cup B) = P(A) + P(B) - P(A\cap B)$
  
* **Monoton铆a** Si $A\subseteq B$ entonces $P(A)\leq P(B)$

* **Probabilidad Condicional** $P(A|B) = \frac{P(A\cap B)}{P(B)}$ si $P(B)>0$

##  Variables Aleatorias y funciones de probabilidad 

#### Variable Aleatoria Discreta

Una variable aleatoria es *discreta* si su conjunto de posibles valores es finito o numerable (por ejemplo, $\mathbb{N}$).

Se caracteriza por una funci贸n de *masa de probabilidad (pmf)* tal que: 

* $p(x)=P(X=x)$ 
  
* $p(x)\geq 0$

* $\sum p(x) = 1$

#### Variable Aleatoria Continua

Una variable aleatoria es *continua* si puede tomar un n煤mero infinito no numerable de valores en un intervalo de $\mathbb{R}$.

Se caracteriza por una funci贸n de *densidad de probabilidad (fdp)* $f(x)$ tal que:

* $f(x) \geq 0$
  
* $\int_{-\infty}^{+\infty}f(x)dx = 1$
  
* $P(a\leq X \leq b) = \int_{a}^{b}f(x)dx$

#### Momentos y FGM

Los *Momentos*, resumen caracter铆sticas importantes de la distribuci贸n de una variable aleatoria. Se clasifican en: 

* Primer Momento : Esperanza Matem谩tica (promedio esperado de $X$)
  
\begin{equation*}
\mathbb{E}(X) = \mu =\begin{cases} \sum_{x}x\cdot p(x) & \textsf{si } X \textsf{es discreta} \\
\int_{-\infty}^{+\infty}x\cdot f(x)dx & \textsf{si } X \textsf{ es continua} \end{cases}
\end{equation*}

* Segundo Momento : Varianza 

\begin{equation*}
\mathbb{Var}(X) = \mathbb{E}[(X-\mu)^2] =\begin{cases} \sum_{x}(x-\mu)^2 \cdot p(x) & \textsf{si } X \textsf{es discreta} \\
\int_{-\infty}^{+\infty}(x-\mu)^2 \cdot f(x)dx & \textsf{si } X \textsf{ es continua} \end{cases}
\end{equation*}


Sea $X$ una variable aleatoria. La *Funci贸n Generadora de Momentos (FGM)* de $X$, denotada por: 

\begin{equation*}
M_{X}(t) = \mathbb{E}[e^{tX}]
\end{equation*}

se define para aquellos valores de $t$ en los que esta esperanza existe (es decir, converge).

**驴C贸mo puedo obtener los momentos a partir de FGM?**

\begin{equation*}
\mu_{k}' = M_{X}(t) = \mathbb{E}[e^{tX}]
\end{equation*}

Esto, significa que el $k$-esimo momento respecto al origen es simplemente la $k$-esima derivada de la FGM evaluada en $t=0$, por ejemplo :

* **`Media`** $\mu = \mathbb{E}[X] = M_{X}(0)'$ 
  
* **`Varianza`** $\sigma^2 = \mathbb{E}[X^2] - \mathbb{E}[X]^2 = M_{X}(0)'' - (M_{X}(0)')^2$ 

* **`Asimetria`** $\mu_3 = \mathbb{E}[(X-\mu)^3] = \mathbb{E}[X^3]-3\mu\mathbb{E}[X^2] + 2\mu^3$
  
* **`Kurtosis`** $\mu_4 = \mathbb{E}[X^4] - 4\mu\mathbb{E}[X^3]+6\mu\mathbb{E}[X^2]-3\mu^4$

### Proceso de Muestreo

Una *muestra* (con reemplazo) consiste en obtener, mediante mecanismos aleatorios, elementos de la poblaci贸n que entran a formar parte de la muestra pero de tal manera que, cuando observamos la caracter铆stica que estamos investigando del primer elemento seleccionado, se devuelve el elemento a la poblaci贸n. Luego se selecciona el segundo elemento de entre todos los elementos, se anota la caracter铆stica y se devuelve a la poblaci贸n, as铆 sucesivamente.

**驴 C贸mo Muestreo ?**

* Sea $X$ una caracter铆stica medible y $f(x,\theta)$ la funci贸n de densidad de probabilidad de la poblaci贸n. 
  
* Se dise帽a un experimento para proporcionar la observaci贸n $X_1$ de la caracter铆stica medible $X$. El experimento se repite bajo las mismas condiciones proporcionando el valor $X_2$. El proceso contin煤a hasta tener $n$ observaciones $X_1, X_2,\cdots, X_n$ de la caracter铆stica $X$. 
  
* Se extrae un elemento y se observa la caracter铆stica medible, llamemosle $X_1$, el objeto se regresa a la poblaci贸n y se mezcla, luego se extrae un segundo elemento, $X_2$ constituye a la segunda observaci贸n. El proceso contin煤a hasta haber extra铆do $n$ elementos para tener una muestra de observaciones $X_1,X_2,\cdots, X_n$ de la caracter铆stica $X$.

**Ejemplo:** Sea una poblaci贸n de 12 alumnos de la clase de Estad铆stica Aplicada en donde la variable aleatoria $X$ representa a la edad del estudiante seleccionado.

* Obtenga una muestra de $n=2$

```python
# Vector de probabilidades y valores
p = np.array([1/12, 6/12, 5/12])
x = np.array([25, 28, 32])

plt.stem(x, p, basefmt = " ")
plt.title("Distribucion de Probabilidad de X")
plt.xlabel("Edad (X)")
plt.ylabel("Probabilidad")
plt.show()
```

```python
# Media Poblacional 
mu = np.sum(p * x)

# Varianza Poblacional
sigma2 = np.sum(p * (x - mu)**2)

# Desviacion estandar 
sigma = np.sqrt(sigma2)

# Muestra 
n_sample = 2 

np.random.choice(x, size = n_sample, replace = True, p = p)
```

Las observaciones $X_1, X_2, \cdots, X_n$ constituyen un conjunto de variables aleatorias independientes dado que el proceso de reemplazo, ninguna observaci贸n se ve afectada por la otra. 

Si las variables aleatorias (*v.a*)  $X_1, X_2, \cdots, X_n$ tienen la misma funci贸n de densidad que la distribuci贸n de la poblaci贸n y su funci贸n de distribuci贸n conjunta, es igual al producto de las marginales, entonces  $X_1, X_2, \cdots, X_n$ forman un conjunto de $n$ variables aleatorias que constituyen la muestra aleatoria de la poblaci贸n. 

### Par谩metro y Estad铆stico

* **Par谩metro** es una caracter铆stica num茅rica del a distribuci贸n de la poblaci贸n de manera que describe, parcial o completamente, la funci贸n de densidad de la caracter铆stica de inter茅s.

* **Estad铆stico** Es cualquier funci贸n real de las variables aleatorias que integran la muestra. Es decir, es una funci贸n de las observaciones muestrales, la cual no contiene ning煤n valor o par谩metro desconocido. En general, un estad铆stico se representa como: 

\begin{equation*}

\end{equation*}

En una **poblaci贸n finita**, de tama帽o $N$, los par谩metros poblacionales media y varianza, vienen dados por:

\begin{equation*}
\mu = \frac{1}{N}\sum_{i=1}^{N}X_i \quad \sigma^2 = \frac{1}{N}\sum_{i=1}^{N}(X_i -\mu)^2
\end{equation*}

Para una **muestra** aleatoria simple de tamano $n$, los **estad铆sticos** media y varianza muestral se definen como: 

\begin{equation*}
\overline{X} = \frac{1}{n}\sum_{i=1}^{n}X_i \quad S^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i -\overline{X})^2
\end{equation*}

### Estimadores y Propiedades

#### Insesgamiento 

Cuando realizamos una estimaci贸n de un par谩metro $\theta$ mediante un estimador estad铆stico $T$, a veces cometemos errores. Estos errores pueden ser por exceso o por defecto. El error, tambi茅n es una variable aleatoria definida como $T-\theta$. La primera propiedad que vamos a requerir a un estimador, es que en t茅rminos medios, el error que se cometa al estimar un par谩metro, sea cero. 

**Definici贸n 1:** Sea $X$ una variable aleatoria y $f(x; \theta)$ su funci贸n de densidad. Diremos que $T$ es un estimador insesgado de $\theta$ si se verifica que $\mathbb{E}(T) = \theta$

**Definici贸n 2:** La diferencia entre la esperanza del estimador y el par谩metro, se llama sesgo del estimador y lo escribiremos como $b(\theta) = \mathbb{E}(T) - \theta$

Si el sesgo es un estimador positivo, este tender谩 a sobreestimar el valor verdadero del par谩metro, sin embargo, si el sesgo es negativo, el estimador tender谩 a infravalorar el valor a estimar. 

**Ejemplo:** Consideremos los tres estimadores de la media de una poblaci贸n normal $\mathcal{N}(2,\sigma^2)$:

\begin{equation*}
\begin{aligned}
\mathbb{E}(\mu_1) &= \frac{n\mu}{(n-1)} \\
\mathbb{E}(\mu_2) &= \mu \\
\mathbb{E}(\mu_3) &=  \frac{n\mu}{(n+1)}
\end{aligned}
\end{equation*}

```python
# Par谩metros
n = 5000
mu = 2
mu1 = n * mu / (n - 100)
mu2 = mu
mu3 = n * mu / (n + 100)

# Simulaci贸n
E1 = np.random.normal(loc = mu1 + 2, size = n)
E2 = np.random.normal(loc = mu2, size = n)
E3 = np.random.normal(loc = mu3 - 2, size = n)

# Crear DataFrame
df = pd.DataFrame({'E1': E1, 'E2': E2, 'E3': E3})

df_long = df.melt(var_name='variable', value_name='value')

plt.figure(figsize = (7, 5))
sns.histplot(data = df_long, x = 'value', hue = 'variable', binwidth = 0.1, 
             element = "step", stat = "count", fill = True, alpha = 0.2, 
             common_norm = False)
plt.xlabel('$\mu$')
plt.ylabel('Frecuencia Absoluta')
plt.tight_layout()
```

El primer estimador, tiene sesgo positivo y tender谩 a sobreestimar la media de la poblaci贸n, mientras que el tercero, tiene sesgo negativo y tender谩 a asignar a la media valores inferiores a su valor real. Frente a este escenario, el estimador $\mu$ es un estimador insesgado.

**Definici贸n 3:** Se llama error cuadr谩tico medio al estimador : 

\begin{equation*}
\textsf{ECM}(T) = \mathbb{E}(T-\theta)^2
\end{equation*}

De la expresi贸n anterior, deduciremos propiedades importantes que debe cumplir un estimador para ser considerado aceptable. 

\begin{equation*}
\textsf{ECM}(T) = \textsf{Var}(T) + (\theta-\mathbb{E}(T))^2
\end{equation*}

* **Contribucion 1.** El tama帽o del error viene dado por la varianza del estimador, es decir, su precisi贸n. Si el estimador tiene poca capacidad de variaci贸n para las distintas muestras que podamos tomar, esto contribuir谩 de forma positiva al obtenci贸n de un error mas peque帽o.

* **Contribucion 2.** El tama帽o del error vendr谩 determinado por la diferencia entre el valor medio que tome el estimador y el par谩metro desconocido. As铆, por ejemplo, si para distintos valores muestrales la medida del estimador coincide con $\theta$ habremos obtenido un buen estimador. 
  
**Ejemplo:** Sea $X\sim\mathcal{N}(\mu, 30)$. Obtenga la esperanza y varianza de los siguiente estimadores propuestos y obtenga el $\textsf{ECM}$. 驴Qu茅 puede concluir con sus resultados?.

\begin{equation*}
\begin{aligned}
\mu_1 &= \frac{1}{n-1}\sum_{i=1}^{n}x_i \\
\mu_2 &= \frac{1}{n}\sum_{i=1}^{n}x_i \\
\mu_3 &= \frac{1}{n+1}\sum_{i=1}^{n}x_i
\end{aligned}
\end{equation*}

```python
# Rango de valores para mu (la media poblacional)
mu_vals = np.linspace(-11, 11, 100)
n = 3
sig2 = 30  # varianza poblacional

# Inicializar arrays para ECM
ECM1, ECM2, ECM3 = [], [], []

# Calcular ECM para cada valor de mu
for mu in mu_vals:
    # Estimadores
    mu1 = (n * mu) / (n - 1)
    mu2 = mu
    mu3 = (n * mu) / (n + 1)
    
    # Varianzas
    var1 = sig2 * n / (n - 1)**2
    var2 = sig2 / n
    var3 = sig2 * n / (n + 1)**2
    
    # Sesgos
    bias1 = mu1 - mu
    bias2 = mu2 - mu  # = 0
    bias3 = mu3 - mu
    
    # ECM = Varianza + (Sesgo)^2
    ECM1.append(var1 + bias1**2)
    ECM2.append(var2 + bias2**2)
    ECM3.append(var3 + bias3**2)

plt.figure(figsize = (7, 5))
plt.plot(mu_vals, ECM1, label=r'$\mu_1 = \frac{1}{n-1}\sum x_i$', color='red', linestyle='--')
plt.plot(mu_vals, ECM2, label=r'$\mu_2 = \frac{1}{n}\sum x_i$', color='blue')
plt.plot(mu_vals, ECM3, label=r'$\mu_3 = \frac{1}{n+1}\sum x_i$', color='green', linestyle=':')

# T铆tulos y etiquetas con LaTeX
plt.xlabel(r'$\mu$', fontsize=14)
plt.ylabel(r'$ECM(\hat{\mu})$', fontsize=14)
plt.title('Error Cuadr谩tico Medio de $\mu$', fontsize=13)

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Eficiencia

Una vez obtenidos dos estimadores insesgados de un parametro, para determinar cual de los dos es mas adecuado debemos prestar atencion a su varianza. Entre los dos elegiremos a aquel que presente menor variacion entorno al valor medio, ya que sera un estimador mas preciso.

* **Definicion 4:** Sea $X$ una variable aleatoria y sea $f(x;\theta)$. Sean $T_1$ y $T_2$ dos estimadores insesgados de $\theta$. Se dice que $T_1$ es mas eficiente que $T_2$ si:
  
\begin{equation*}
\textsf{Var}(T_1) < \textsf{Var}(T_2) 
\end{equation*}

* **Definicion 5:** Sea $X$ una variable aleatoria y sea $f(x;\theta)$ su funci贸n de densidad. Si $T$ es un estimador insesgado de $\theta$ y no hay ningun otro estimador que tenga minima varianza, entonces se dice que $T$ es el estimador insesgado de minima varianza de $\theta$. 

Estas dos propiedades, las podemos relacionar con ECM de un estimador.

\begin{equation*}
\textsf{ECM}(T) = \textsf{Var}(T) + (\theta - \mathbb{E}(T))^2
\end{equation*}

De tal forma que entre todos los estimadores insesgados, aquel que presente una varianza minima sera el que presente un error cuadratico minimo. 驴 Podremos encontrar un estimador sesgado con menor $\textsf{ECM}$ ?. El problema de obtener estimadores con varianza minima, esta solucionado por el siguiente resultado.

* **Definicion 6:** Sea $X$ una variable aleatoria y sea $f(x;\theta)$ su funci贸n de densidad. Tomemos una muestra aleatoria simple de tama帽o $n$ y sea $T$ un estimador de $\theta$. Entonces : 
  
\begin{equation*}
\textsf{Var}(T) \geq \frac{(1+b'(\theta))^2}{n\mathbb{E}\left(\frac{\partial\ln f(x;\theta)}{\partial\theta}\right)^2}
\end{equation*}

Donde $+b'(\theta)$ es la primera derivada del sesgo del estimador. La cota de Cramer-Rao permite determinar el valor m铆nimo que puede alcanzar la varianza de un estimador de un par谩metro desconocido. 

#### Consistencia

Las dos propiedades anteriores, tienen vigencia independiente cual sea el tama帽o muestral. El factor *tama帽o muestral* debe tambi茅n ayudarnos a determinar la calidad del estimador. Parece l贸gico exigir a un buen estimador que al aumentar el tama帽o muestral, su precisi贸n aumente.

* **Definici贸n 7:** Sea $X$ una variable aleatoria y sea $f(x;\theta)$ su funci贸n de densidad. Sea $T_n$ es un estimador de $\theta$ para una muesta de tama帽o $n$. Diremos que $T_n$ es un estimador consistente si la sucesi贸n de estimadores $\{T_n\}$ converge en probabilidad a $\theta$, es decir: 

\begin{equation*}
\lim_{n\to\infty}\textsf{Prob}(|T_n - \theta|\geq \epsilon) = 0 \quad\forall\epsilon>0
\end{equation*}

Una caracter铆stica de los estimadores consistentes que resulta m谩s simple de aplicar en la pr谩ctica, es la siguiente.

* **Proposici贸n:**  Sea $X$ una variable aleatoria y sea $f(x;\theta)$ su funci贸n de densidad. Adem谩s, sea $T_n$ un estimador de $\theta$ para una muestra de tama帽o $n$. Si se verifica que: 

\begin{equation*}
\lim_{n\to\infty}\mathbb{E}(T_n) = \theta\quad \lim_{n\to\infty}\textsf{Var}(T_n)=0
\end{equation*}

De esta manera, un estimador consistente ser谩 aquel que, a medida que la muestra aumenta, el valor esperado del estimador tiende hacia al verdadero valor del par谩metro (se vuelve insesgado) y la varianza del estimador se hace despreciable (la varianza tiende a 0). 

\begin{equation*}
\textsf{ECM}(T) = \underbrace{\textsf{Var}(T)}_{\lim_{n\to\infty}\textsf{Var}(T_n)=0} + \underbrace{(\theta - \mathbb{E}(T))^2}_{\lim_{n\to\infty}\mathbb{E}(T_n) = \theta}
\end{equation*}

```python
# Par谩metros
n_sim = 10000  # n煤mero de lanzamientos
true_mean = 0.5  # media te贸rica de una Bernoulli(0.5)

# Simulaci贸n: lanzamientos de moneda (Bernoulli con p=0.5)
samples = np.random.binomial(n = 1, p = 0.5, size = n_sim)

# Media acumulada
sample_means = np.cumsum(samples) / np.arange(1, n_sim + 1)

# Gr谩fico
plt.figure(figsize=(7, 5))
plt.plot(sample_means, label = 'Media muestral', color = 'blue')
plt.axhline(y = true_mean, color = 'red', linestyle='--', label = 'Media poblacional (0.5)')
plt.title('Ley de los Grandes N煤meros - Lanzamientos de Moneda')
plt.xlabel('N煤mero de lanzamientos')
plt.ylabel('Proporci贸n de caras (media acumulada)')
plt.legend()
plt.tight_layout()
plt.show()
```

As铆, las estimaciones pueden ser de dos tipos:

* **`Estimaci贸n por Intervalos`** Proporciona un rango de posibles valores, dentro cual estimamos que estar谩 el par谩metro desconocido.

Un *estimador puntual* es un *estad铆stico muestral* (media aritm茅tica, varianza, etc) que se emplea para estimar par谩metros (media poblacional, varianza poblacional, etc). * **Definici贸n 8:** Un estimador puntual de un par谩metro $\theta$ es un valor que puede ser considerado representativo de $\theta$ y se indicar谩 como $\hat{\theta}$. Se obtuene a partir de la funci贸n de la muestra. 

##### M谩xima Verosimilitud

Sea una muestra aleatoria $\{X_1, X_2, \cdots, X_n\}$ procedente de una poblacion con funcion de probabilidad $P(x;\theta)$ o con funcion de densidad $f(x;\theta)$ donde $\theta$ es un parametro desconocido que toma valores en el espacio parametrico $\Omega$, $\theta\in\Omega$. La funcion de densidad conjunta de una muestra aleatoria la indicaremos por: 

\begin{equation*}
f(x_1, x_2, \cdots, x_n) = \prod_{i=1}^{n}f(x_i ; \theta)
\end{equation*}

* **Definici贸n 9:** La funcion de verosimilitud de $n$ variables aleatorias, se define como: 

\begin{equation*}
L(x;\theta) = L(x_1, x_2, \cdots, x_n; \theta) = \prod_{i=1}^{n}f(x_i ; \theta)
\end{equation*}

* **Definici贸n 10:** El m茅todo de m谩xima verosimilitud consiste en elegir como estimador del par谩metro desconocido $\theta$ aque valor que $\hat{\theta}(x_1, x_2, \cdots, x_n)$ hace maxima la funcion de verosimilitud $L(x_1, x_2, \cdots, x_n; \theta)$. Es decir, consiste en encontrar aquel valor de $\hat{\theta}(x_1, x_2, \cdots, x_n)$ tal que: 
  
\begin{equation*}
 \hat{\theta}(x_1, x_2, \cdots, x_n) = \argmax_{\theta\in\Omega} L(x_1, x_2, \cdots, x_n; \theta)
\end{equation*}

**Ejemplo:** Sea $X_n$ una sucesi贸n de variables aleatorias con funci贸n de probabilidad $P(x_i; \theta)$ tal que

\begin{equation*}
P(X=x) = \theta^{x_i}(1-\theta)^{1-x_i}
\end{equation*}

Encuentre el estimador m谩xima verosimilitud $\hat{\theta}$ para el par谩metro $\theta$.

**Soluci贸n:** 

\begin{equation*}
\mathcal{L}(\theta) = \sum_{i=1}^{n}\left[ x_i \log{\theta} + (1-x_i)\log{(1-\theta)} \right]
\end{equation*}

```python
from scipy.optimize import minimize

n = 500
theta_true = 0.7
x = np.random.binomial(1, theta_true, n)

def neg_log_likelihood(theta): 
    if theta <= 0 or theta>=1:
        return np.inf 
    return -np.sum(x * np.log(theta) + (1-x)*np.log(1 - theta))

result = minimize(neg_log_likelihood, x0 = 5, bounds=[(1e-5, 1 - 1e-5)])
theta_mle = result.x[0]

print(f"Valor verdadero de p: {theta_true}")
print(f"Estimador de m谩xima verosimilitud (MLE): {theta_mle:.4f}")

theta_grid = np.linspace(0.01, 0.99, 200)
log_likelihood = [ -neg_log_likelihood(theta) for theta in theta_grid ]

plt.figure(figsize = (7, 5))
plt.plot(theta_grid, log_likelihood, label='Log-Verosimilitud')
plt.axvline(theta_mle, color = 'red', linestyle = '--', label = f'MLE: p={theta_mle:.4f}')
plt.axvline(theta_true, color = 'green', linestyle = '--', label = f'Valor verdadero: {theta_true}')
plt.xlabel(r'$\theta$', fontsize = 14)
plt.ylabel(r'$\log  \mathcal{L}(\theta)$', fontsize = 14)
plt.title('Funcion de Log-Verosimilitud - Modelo Bernoulli')
plt.legend()
plt.show()
```

**Ejemplo:** Sea $X_n$ una sucesi贸n de variables aleatorias con funci贸n de probabilidad $P(x_i ; \lambda)$ tal que: 

\begin{equation*}
 P(X=x) = e^{-\lambda}\frac{\lambda^x_i}{x!}
\end{equation*}

Encuentre el estimador de m谩xima verosimilitud $\hat{\lambda}$ para el par谩metro $\lambda$. 

**Soluci贸n:** 

\begin{equation*}
\mathcal{L}(\lambda) = -n\lambda + \left(\sum_{i=1}^{n}x_i\right)\log{\lambda} - \sum_{i=1}^{n}\log{x_i !}
\end{equation*}

```python
from scipy.special import gammaln 

lambda_true = 4.5
x = np.random.poisson(lam = lambda_true, size = n)

def neg_log_likelihood(lmbda):
    if lmbda <= 0:
        return np.inf
    return -(-n * lmbda + np.sum(x) * np.log(lmbda) - np.sum(gammaln(x + 1)))

result = minimize(neg_log_likelihood, x0=1.0, bounds=[(1e-5, None)])
lambda_mle = result.x[0]

print(f"Valor verdadero de lambda: {lambda_true}")
print(f"Estimador MLE: {lambda_mle:.4f}")
```

```python
lambda_grid = np.linspace(1, 8, 200)
log_likelihood = [-neg_log_likelihood(lmbda) for lmbda in lambda_grid]

plt.figure(figsize = (7, 5))
plt.plot(lambda_grid, log_likelihood, label = 'Log-Verosimilitud')
plt.axvline(lambda_mle, color = 'red', linestyle = '--', label = f'MLE: 位 = {lambda_mle:.2f}')
plt.axvline(lambda_true, color = 'green', linestyle = '--', label = f'Valor real: 位 = {lambda_true}')
plt.xlabel(r'$\lambda$', fontsize = 14)
plt.ylabel(r'$\log \mathcal{L}(\lambda)$', fontsize = 14)
plt.title('Funcion Log-Verosimilitud - Modelo Poisson')
plt.legend()
plt.show()
```
