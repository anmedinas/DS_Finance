### **1.2 Distribuciones Discretas**

#### **1.2.1 Distribuci칩n Bernoulli $B(p)$**

Un experimento aleatorio tiene una distribuci칩n de probabilidad:

\begin{equation*}
f(x) = p^x(1-p)^{(1-x)}\quad x\in\{ 0,1\}
\end{equation*}

donde hay dos posibles resultados:

* 0, con probabilidad $1-p$
* 1, con probabilidad $p$

Cualquier experimento aleatorio con dos resultados se puede representar mediante una distribuci칩n de Bernouilli asignando a uno de los dos posibles resultados el valor 0 y al otro, el valor 1.

El siguiente c칩digo, genera variables aleatorias distribuidas bernoulli y adicionalmente, presenta un histograma.

```python
from scipy.stats import bernoulli
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import warnings

# Configuracion de Seaborn
sns.set_style('darkgrid')
sns.set_theme(rc={'figure.figsize':(5.27,5.27)})

# Omite warnings por desactualizacion u otro
warnings.simplefilter("ignore")

# Semilla aleatoria
np.random.seed(20210730)

# Generacion de datos bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)

# Visualizacion distribucion
ax = sns.histplot(data_bern, kde = False, color = "skyblue")
ax.set(xlabel='x', ylabel='Frequencia',title="Histograma v.a Bernoulli")
plt.show()
```

#### **1.2.2 Distribuci칩n Binomial $B(n,p)$**

Un experimento binomial, tiene una distribuci칩n de probabilidad:

\begin{equation}
f(x) = p^x(1-p)^{(n-x)}\quad x\in\{ 0,1,2,\cdots\}
\end{equation}

Esta funci칩n se debe a que dado que tenemos un experimiento bernoulli con probabilidad $p$, el experimento se repite $n$ veces de forma independiente. Los par치metros se definen como:

* $p$ : probabilidad de ganar cada juego individual.
* $n$ : el n칰mero total de veces en que jugamos.

```python
from scipy.stats import binom
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import warnings

# Configuracion de Seaborn
sns.set_style('darkgrid')
sns.set_theme(rc={'figure.figsize':(5.27,5.27)})

# Omite warnings por desactualizacion u otro
warnings.simplefilter("ignore")

# Semilla aleatoria
np.random.seed(20210730)

# Generacion de datos binomiales
data_binom = binom.rvs(n=10, p=0.8, size=10000)

# Visualizacion de Distribucion
ax = sns.histplot(data_binom, kde=False, color='skyblue')
ax.set(xlabel='x', ylabel='Frequencia', title="Histograma v.a Binomial")
plt.show()
```

La funci칩n de distribuci칩n geom칠trica, tiene la siguiente distribuci칩n de probabilidad:

La distribuci칩n geom칠trica es un modelo adecuado para aquellos procesos en los que se repiten pruebas hasta la consecuci칩n del 칠xito a resultado deseado y tiene interesantes aplicaciones en los muestreos realizados de esta manera . Tambi칠n implica la existencia de una dicotom칤a de posibles resultados y la independencia de las pruebas entre s칤.

#### **1.2.4 Distribuci칩n Poisson $P(\mu)$**

La funci칩n de probabilidad poisson, tiene la siguiente expresi칩n:

\begin{equation}
p(r;\mu) = \frac{\mu^r e^{-\mu}}{r!}\quad\mu\in\{0,\infty\}\quad r\in\{0,1,2,3,\cdots \}
\end{equation}

Esta funci칩n, expresa, a partir de una frecuencia de ocurrencia media (tasa de ocurrencia), la probabilidad de que ocurra un determinado n칰mero de eventos durante cierto per칤odo de tiempo. Concretamente, se especializa en la probabilidad de ocurrencia de sucesos con probabilidades muy peque침as, o sucesos ``raros``.

#### **1.3.2 Distribuci칩n Normal $\mathcal{N}(\mu,\sigma^2)$**

Dada una variable aleatoria $X$, decimos que la frecuencia de sus observaciones puede aproximarse satisfactoriamente a una normal, donde la funci칩n de densidad de probabilidad tiene el siguiente aspecto.

\begin{equation*}
f(x) = \frac{1}{\sqrt{2\pi\sigma}}\exp{\left( -\frac{(x-\mu)^2}{2\sigma^2}\right)}\quad  \mu,x\in\mathbb{R}\quad\sigma > 0
\end{equation*}

Donde, los par치metros de esta distribuci칩n, son la media o par치metro de localizaci칩n $\mu$ y la desviaci칩n t칤pica o par치metro de escala $\sigma$.

```python
from scipy.stats import norm
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import warnings

# Configuracion de Seaborn
sns.set_style('darkgrid')
sns.set_theme(rc={'figure.figsize':(5.27,5.27)})

# Omite warnings por desactualizacion u otro
warnings.simplefilter("ignore")

# Semilla aleatoria
np.random.seed(20210730)

# Generacion de datos normales
data_normal = norm.rvs(size=10000, loc=0, scale=1)

# Visualizacion distribucion
ax = sns.histplot(data_normal,bins = 100, kde=True, color='skyblue')
ax.set(xlabel="x", ylabel="Frecuencia", title="Histograma v.a Normal")
plt.show()
```

#### **1.3.4 Distribuci칩n $t$-Student $t(v)$**



Sea $X$ una variable aleatoria continua y $v>0$, si $X$ tiene una distribuci칩n $t$ con $v$ grados de libertad, entonces la funci칩n de densidad viene por la siguiente expresi칩n.

#### **1.3.6 Distribuci칩n Gamma $\Gamma(\alpha,\lambda)$**

Si una variable aleatoria continua $X$, tiene distribuci칩n gamma con par치metros $\alpha>0$ y
$\lambda > 0$ entonces, la funci칩n de densidad viene dada por la siguiente expresi칩n.

\begin{equation*}
f(x) = \frac{\lambda(\lambda x)^{\alpha-1}\exp{(-\lambda x)}}{\Gamma(\alpha)} \quad \alpha>0,\lambda>0\quad x\in\left(0,\infty \right)
\end{equation*}

Donde $\Gamma(\alpha)$ es la funci칩n gamma y se define como:

\begin{equation*}
\Gamma(\alpha) = \int_{0}^{\infty} t^{\alpha-1}\exp{(-t)}dt
\end{equation*}

#### **1.3.7 Distribuci칩n Beta $\mathcal{Be}(\alpha,\beta)$**

Si una variable aleatoria $X$ continua tiene distribuci칩n beta con par치metros $\alpha,\beta>0$ entonces la funci칩n de densidad viene dada por:

\begin{equation*}
f(x) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}\quad 0<x<1
\end{equation*}

Donde $B(\alpha,\beta)$ es la funci칩n beta y se define para $\alpha,\beta>0$ como:

\begin{equation*}
B(\alpha,\beta) = \int_{0}^{1} x^{\alpha-1}(1-x)^{\beta-1}dx
\end{equation*}

## **2. Ejercicios Propuestos de Simulaci칩n Estadistica** 游빏

**Definici칩n**: Si $X$ es una variable aleatoria continua con *cdf*
$F_{X}(x)$, entonces $X=F_{X}(x) \sim U(0,1)$. El m칠todo de transformaci칩n
inversa para generar variables aleatorias aplica la transformaci칩n de probabilidad
como una integral. Definamos la transformaci칩n inversa como:

\begin{equation*}
F_{X}^{-1} = \inf\{x:F_{X}(x)=u\},\quad\quad 0 < u < 1
\end{equation*}

Si $U\sim\mathcal{U}(0,1)$, entonces, para todo $x\in\mathbb{R}$

\begin{equation*}
\begin{aligned}
P(F_{X}^{-1}\leq x) &= P(\inf\{t:F_{X}(t)=U \}\leq x)\\
  &= P(U\leq F_{X}(x))\\
  &= F_{U}(F_{X}(x))\\
  &= F_{X}(x)
\end{aligned}
\end{equation*}

**Propuesto 1:** Utilizando la definici칩n anterior, genere 100.000
*v.a* a partir de la  distribuci칩n exponencial de tasa $\lambda=2$ y compare resultados mediante simulaci칩n mediante la funci칩n ``expon`` de scipy.


```python
from scipy.stats import uniform, expon
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import warnings

# Omite warnings por desactualizacion u otro
warnings.simplefilter("ignore")

# Semilla Aleatoria
np.random.seed(20210730)

# Parametros
n  = 100000
mu = 2
theta = 1.0/mu

# Generacion de variable aleatoria transformada
u  = uniform.rvs(size=n, loc=0, scale=1)
x  = -theta *  np.log(1-u)

# Generacion de variable aleatoria exponencial
d  = expon.rvs(scale=theta, loc=0, size=n)

# Visualizacion distribucion
ax = sns.histplot(x,bins = 100, kde=False, color='green')
ax.set(xlabel="x", ylabel="Frecuencia", title="Histograma Variable Transformada")
plt.show()
```

```python
print("La media de la variable simulada es: ", np.mean(x).round(4), "y la varianza es: ", np.var(x).round(4))
print("La media de la variable real es: ", np.mean(d).round(4), "y la varianza es: ", np.var(d).round(4))
```

```python
from scipy import optimize
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns

# Configuracion de Seaborn
sns.set_style('darkgrid')
sns.set_theme(rc={'figure.figsize':(5.27,5.27)})

# Omite warnings por desactualizacion u otro
warnings.simplefilter("ignore")

# Construyo la funcion p(x) para visualizar
def p(x):
    return (2*np.exp(-(x-2)**2) + np.exp(-np.abs(x)) + np.exp(-(x+2)**2)*np.exp(-np.abs(x+2)))

# Construyo la funcion p1(x) para optimizar
def p1(x):
    return -(2*np.exp(-(x-2)**2) + np.exp(-np.abs(x)) + np.exp(-(x+2)**2)*np.exp(-np.abs(x+2)))

# Genero datos para plotear
x = np.linspace(-10, 10, 1000)
y = p(x)

# Punto inicial
x0 = -3

# Grafico funcion
fig, ax = plt.subplots(figsize=(5,5))
plt.plot(x,y,'-')
ax.set_title('$p(x) \propto 2e^{-(x-2)^2} + e^{-|x|} + e^{-(x+2)^2}e^{-|x+2|}$', fontsize=15)
ax.set_xlabel('$x$', fontsize = 13)
ax.set_ylabel('$p(x)$', fontsize = 13)
ax.scatter(x0, p(x0), marker='o', s=300)
plt.show()
```

**Propuesto 6**: Sea el Modelo lineal gausiano


\begin{equation*}
Y = X\beta + \epsilon
\end{equation*}

Luego, la media y varianza quedan caracterizadas como:


\begin{equation*}
\begin{aligned}
\mathbb{E}(Y|X) &= \mathbb{E}(X\beta + \epsilon|X) = \mathbb{E}(X\beta|X)=X\beta \\
\mathbb{V}(Y|X) &= \mathbb{V}(X\beta + \epsilon|X) = \mathbb{V}(\epsilon|X)=\sigma^2 I
\end{aligned}
\end{equation*}


El proceso generador de datos y la distribuci칩n de los estimadores vienen dados por:

\begin{equation*}
\begin{aligned}
Y|X &\sim \mathcal{N}(X\beta,\sigma^2 I) \\
\hat{\beta}|X &\sim \mathcal{N}(\beta,\sigma^2(X^{T}X)^{-1})
\end{aligned}
\end{equation*}

Implemente computacionalmente la log-verosimilitud del modelo de regresi칩n lineal y corrobore los estimadores con los obtenidos mediante statsmodels.

Recordar que el MLE de un modelo de regresi칩n lineal, es de la siguiente forma:

\begin{equation*}
\mathcal{l}(\beta,\sigma^2) =
\sum_{i=1}^{n}\log\left(\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left[-\frac{(y_{i}-x_{i}^{\top}\beta)^2}{2\sigma^2}\right]\right)
\end{equation*}




```python
from scipy.optimize import minimize
from scipy.stats import norm
import scipy.stats as stats
import numpy as np
import random
import matplotlib.pyplot as plt

# Semilla aleatoria
np.random.seed(20210730)

# Parametros de inicializacion
N = 100
beta_0 = 3
beta_1 = 8

# Generacion de muestras aleatorias
X = np.random.normal(loc = 10, scale = 2, size = N)
e = np.random.normal(loc = 0, scale = 2, size = len(X))
Y = beta_0 + beta_1 * X + e
```

```python
def log_lik(par_vec, y, x):
    # Si la desviacion estandar es negativa, devuelve un numero grande
    if par_vec[2] < 0:
        return(1e8)
    # La funcion de verosimilitud se define como:
    lik = norm.pdf(y, loc = par_vec[0] + par_vec[1] * x, scale = par_vec[2])
    # Esto es similar a calcular la verosimilitud de  Y - XB
    # res = y - par_vec[0] - par_vec[1] * x
    # lik = norm.pdf(res, loc = 0, sd = par_vec[2])

    # Si todos los logaritmos son ceros, retorna un valor grande
    if all(v == 0 for v in lik):
        return(1e8)
    # Logaritmo de cero = -Inf
    return(-sum(np.log(lik[np.nonzero(lik)])))
```

```python
import scipy.optimize as optimize
opt_res = minimize(fun = log_lik, x0 = [0, 0, 10], args = (Y, X))
print(opt_res)
```
